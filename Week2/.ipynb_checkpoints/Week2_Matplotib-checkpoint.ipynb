```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Sine graph
x = np.arange(0, 10, 0.1) #[0, 0.1, 0.2 ..]
y = np.sin(x)

#Graph creation
plt.plot(x,y)

#Determine chart properiteis
plt.title("Sine graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

#Show he graph
plt.show()
```

```python
#Point chart
x = np.random.rand(50)
y = np.random.rand(50)
colers = np.random.rand(50)
sizes = np.random.randint(50, 150, 50)

#Graph creation
plt.scatter(x, y, c=colers, s=sizes)

#properiteis
plt.title("Point chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")

#Show he graph
plt.show()
```

```python
#Hisogram
data = np.random.rand(1000) #Generate random numbers

#Graph creation
plt.hist(data, bins=30) #Numbers of bars

#properiteis
plt.title("Hisogram")
plt.xlabel("X axis")
plt.ylabel("Y axis")

#Show he graph
plt.show()
```

```python
#Column chart
x = ['A', 'B', 'C', 'D', 'E']
y = np.random.randint(1,10,5)

#Graph creation
plt.bar(x, y)

#properiteis
plt.title("Column chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")

#Show he graph
plt.show()
```

```python
#Pie chart
sizes = [10, 10, 25, 10, 15, 5]

#Graph creation
plt.pie(sizes)

#properiteis
plt.title("Pie chart")

#Show he graph
plt.show()
```
